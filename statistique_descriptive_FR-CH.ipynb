```python
import pandas as pd
import pyreadstat
import numpy as np
import json
pd.set_option('display.max_columns', None)
```

```python
df_chine = pd.read_csv("/Users/fnacboulogne/Desktop/app quanti covid/app/QuantiCovid_isolationForestCH.csv")
df_chine.describe()
```

```python
df_france = pd.read_csv("/Users/fnacboulogne/Desktop/app quanti covid/app/QuantiCovid_isolationForestFR.csv")
df_france.describe()
```

```python
df_reshaped_france = df_france.copy()
drop_indices = np.random.choice(df_reshaped_france.index, df_reshaped_france.shape[0] - df_chine.shape[0], replace=False)
df_reshaped_france = df_reshaped_france.drop(drop_indices)
```

```python
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency

column_to_check = []
for column in df_chine.columns:
    if column in df_france.columns:
        if df_france[column].dtype != 'O' :
            
            stat, p_value = ttest_ind(df_chine[column] ,df_reshaped_france[column])
            data_test = pd.DataFrame()
            data_test["1"] = df_chine[column]
            data_test["2"] = df_reshaped_france[column]
            
            cont = data_test[["1", "2"]].pivot_table(index="1", columns="2", aggfunc=len).fillna(0).copy().astype(int)
            st_chi2, st_p, st_dof, st_exp = chi2_contingency(cont)

            if p_value < 0.05 and st_p < 0.05:
                column_to_check.append(column)
```

```python
len(column_to_check)
```

```python
df_col = pd.read_excel("/Users/fnacboulogne/Downloads/donnée_Quanti_avecMeta.xlsx", sheet_name="MetaColumn")
df_col = df_col.replace(' ', '_', regex=True)
df_col.head()
```

```python
with open('/Users/fnacboulogne/Downloads/QuantiMetadata.json') as json_file:
    dict_metadata = json.load(json_file)
```

```python
df_meta_fr = df_france.copy()
jsoneq = ""
for columns in df_meta_fr:
    for head in df_col :
        if df_col[head].iloc[1] == columns:
            jsoneq = df_col[head].iloc[0]
            break

    if jsoneq in dict_metadata :
        if df_meta_fr[columns].dtype != 'O' :
            df_meta_fr[columns] = df_meta_fr[columns].astype(int, errors='ignore').astype(str).map(dict_metadata[jsoneq]).fillna(df_france[columns])
            
        else :
            df_meta_fr[columns] = df_meta_fr[columns].astype(str).map(dict_metadata[jsoneq]).fillna(df_france[columns])
```

```python
df_meta_ch = df_chine.copy()
jsoneq = ""
for columns in df_meta_ch:
    for head in df_col :
        if df_col[head].iloc[1] == columns:
            jsoneq = df_col[head].iloc[0]
            break
    if jsoneq in dict_metadata :
        if df_meta_ch[columns].dtype != 'O' :
            df_meta_ch[columns] = df_meta_ch[columns].astype(int, errors='ignore').astype(str).map(dict_metadata[jsoneq]).fillna(df_chine[columns])
        else :
            df_meta_ch[columns] = df_meta_ch[columns].astype(str).map(dict_metadata[jsoneq]).fillna(df_chine[columns])
```

```python
import plotly.express as px
i = 0
for column in column_to_check:
    if np.max(df_france[column]) > 0 and np.max(df_chine[column]) > 0 :
        df_meta_fr = df_meta_fr.reindex(df_france.sort_values(by=column).index)
        df_meta_ch = df_meta_ch.reindex(df_chine.sort_values(by=column).index)
        fig_france = px.histogram( df_meta_fr[~df_meta_fr[column].isin(["-1",-1,"-1.0",-1.0])  ][column] , x=column,title='Répartition de {} pour la France :'.format(column), histnorm='probability density')
        fig_france.show()

        fig_chine = px.histogram( df_meta_ch[~df_meta_ch[column].isin(["-1",-1,"-1.0",-1.0])] [column], x=column,title='Répartition de {} pour la Chine :'.format(column),color_discrete_sequence=['indianred'], histnorm='probability density')
        fig_chine.show()
        if i == 10 :
            break
        i+=1
    column_to_check.remove(column)
    
```
