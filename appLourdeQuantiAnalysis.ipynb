```python
def FactorAnalysis_dashboard( ):
    key = selected_factor.get()
    factor = ["componant_Vol_ecologie","componant_Vol_economie","componant_Trend_Local","Componant_Trend_Prix","Componant_Trend_secu","Componant_confiance_marque","Componant_confiance_comp","Componant_confiance_benev","Componant_confiance_integ","Componant_image_marque","Componant_image_val","Componant_reconditionnement","Componant_image_RSE","Componant_changement_internet","Componant_confiance","Componant_image","componant_legitimite","componant_nouveaute","componant_cat_social","componant_habitation"]
    value_selection = [
        ["accord_achat_ecolo_postconf","intention_ecologie_postconf"],["accord_achat_econome_postconf","intention_economie_postconf","intention_economie2_postconf"],
         ["TRENDS_LOC_1" , "TRENDS_LOC_2" , "TRENDS_LOC_3"],["TRENDS_PRIX_1" , "TRENDS_PRIX_2" , "TRENDS_PRIX_3"],["TRENDS_SECU_1" , "TRENDS_SECU_2" , "TRENDS_SECU_3" , "TRENDS_SECU_4"],
        ["confiance_distrib_nAlim" , "confiance_distrib_alim" , "confiance_distrib" , "confiance_brico"],["CONF_COMP_1" , "CONF_COMP_2" , "CONF_COMP_3"],["CONF_BENEV_1" , "CONF_BENEV_2" , "CONF_BENEV_3"],["CONF_INTEG_1" , "CONF_INTEG_2" ,  "CONF_INTEG_3" , "CONF_INTEG_4"],["image_distrib_nAlim" , "image_distrib_alim" , "image_distrib" , "image_brico"],["IMG_VAL_1" , "IMG_VAL_2" , "IMG_VAL_3" , "IMG_VAL_4" ],
         ["raison_achat_reconditionner_postconf" , "achat_magasin_reconditionne_postconf" , "achat_enligne_reconditionne_postconf"], ["IMG_RSE_1" , "IMG_RSE_2" , "IMG_RSE_3" , "IMG_RSE_4" , "IMG_RSE_5"], ["habitude_InternetToPointvente_postconf" , "habitude_InternetToDrive_postconf" , "habitude_InternetTolocker_postconf"],["CONF_COMP_1", "CONF_COMP_2" , "CONF_COMP_3", "CONF_BENEV_1",  "CONF_BENEV_2",  "CONF_BENEV_3", "CONF_INTEG_1", "CONF_INTEG_2",  "CONF_INTEG_3", "CONF_INTEG_4", "confiance_distrib_nAlim", "confiance_distrib_alim", "confiance_distrib", "confiance_brico"],["IMG_VAL_1", "IMG_VAL_2", "IMG_VAL_3",  "IMG_VAL_4" , "IMG_RSE_1" , "IMG_RSE_2", "IMG_RSE_3", "IMG_RSE_4", "IMG_RSE_5", "image_distrib_nAlim", "image_distrib_alim", "image_distrib", "image_brico" ],
        ["LEGIT_1", "LEGIT_2", "LEGIT_3", "LEGIT_4", "LEGIT_5", "LEGIT_6", "LEGIT_7", "LEGIT_9"],["TRENDS_NEW_1", "TRENDS_NEW_2", "TRENDS_NEW_3"]
        ,["Diplome_FR","CSP", "Cat_Revenue_FR"],
        ["cat_Agglo", "cat_Agglo_simple", "cat_zone_habitation"]
    ]
    key = value_selection[factor.index(key)]
    
    frame = Toplevel(root)

    text_box = tk.Text(frame, width = 80, height = 30)
    
    text_box.delete(1.0, tk.END)
    text_box.insert(tk.END, "alpha de crombach : {}".format(pg.cronbach_alpha(data=df[key], ci=.95)))

    X = StandardScaler().fit_transform(df[key])
    feature_names = key

    n_comps = 1

    methods = [
        ("PCA", PCA()),
        ("Unrotated FA", FactorAnalysis()),
        ("Varimax FA", FactorAnalysis(rotation="varimax")),
    ]

    fig, axes = plt.subplots(ncols=len(methods), figsize=(10, 8))

    for ax, (method, fa) in zip(axes, methods):
        fa.set_params(n_components=n_comps)
        fa.fit(X)

        components = fa.components_.T
        text_box.insert(tk.END,"\n\n %s :\n" % method)
        fa_loadings = fa.components_.T    # loadings

        # variance explained
        total_var = X.var(axis=0).sum()  # total variance of original variables,
                                            # equal to no. of vars if they are standardized

        var_exp = np.sum(fa_loadings**2, axis=0)
        prop_var_exp = var_exp/total_var
        cum_prop_var_exp = np.cumsum(var_exp/total_var)
        
        text_box.insert(tk.END, f"variance explained: {var_exp.round(2)} \n")
        text_box.insert(tk.END, f"proportion of variance explained: {prop_var_exp.round(3)} \n")
        text_box.insert(tk.END, f"cumulative proportion of variance explained: {cum_prop_var_exp.round(3)} \n")

        vmax = np.abs(components).max()
        ax.imshow(components, cmap="RdBu_r", vmax=vmax, vmin=-vmax)
        ax.set_yticks(np.arange(len(feature_names)))
        if ax.is_first_col():
            ax.set_yticklabels(feature_names)
        else:
            ax.set_yticklabels([])
        ax.set_title(str(method))
        ax.set_xticks([0])
        ax.set_xticklabels(["Comp. 1"])

    fig.suptitle("Factors")
    plt.tight_layout()
    
    text_box.pack()
    
    canvas = FigureCanvasTkAgg(plt,master = root)       
    canvas.draw() 
    canvas.get_tk_widget().pack() 
    toolbar = NavigationToolbar2Tk(canvas,root) 
    toolbar.update() 
    canvas.get_tk_widget().pack() 
    
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def vif_study():
    key = selected_comp.get()
    factor = ["Changement_Livraison","Changement_Deplacement","Image","confiance","Volonte_Economie","Legitimite","Trend_nouveaute","Achat_Reconditionne"]
    
    value_selection = [ ["Age", "IMG_VAL_1", "IMG_VAL_2", "IMG_VAL_3", "IMG_VAL_4","IMG_RSE_1", "IMG_RSE_2", "IMG_RSE_3", "IMG_RSE_4", "IMG_RSE_5", "accord_consomation_hédoniste_postconf", "accord_consomation_hédoniste2_postconf","cat_Agglo", "cat_Agglo_simple", "cat_zone_habitation"],
                        ["Age", "accord_consomation_hédoniste_postconf", "accord_consomation_hédoniste2_postconf","raison_achat_reconditionner_postconf", "achat_magasin_reconditionne_postconf", "achat_enligne_reconditionne_postconf"],
                        ["CONF_COMP_1", "CONF_COMP_2", "CONF_COMP_3", "CONF_BENEV_1", "CONF_BENEV_2", "CONF_BENEV_3",  "CONF_INTEG_1", "CONF_INTEG_2",  "CONF_INTEG_3", "CONF_INTEG_4"],
                        [ "LEGIT_1", "LEGIT_2", "LEGIT_3", "LEGIT_4", "LEGIT_5", "LEGIT_6", "LEGIT_7", "LEGIT_9"],
                        ["accord_achat_econome_postconf", "intention_economie_postconf", "intention_economie2_postconf"],
                        ["TRENDS_PRIX_1", "TRENDS_PRIX_2", "TRENDS_PRIX_3","TRENDS_LOC_1", "TRENDS_LOC_2", "TRENDS_LOC_3", "TRENDS_SECU_1" , "TRENDS_SECU_2" , "TRENDS_SECU_3" , "TRENDS_SECU_4"],
                        ["IMG_VAL_1", "IMG_VAL_2", "IMG_VAL_3", "IMG_VAL_4","IMG_RSE_1", "IMG_RSE_2", "IMG_RSE_3", "IMG_RSE_4", "IMG_RSE_5"],
                        ["accord_achat_econome_postconf","intention_economie_postconf","intention_economie2_postconf","accord_consomation_hédoniste_postconf","accord_consomation_hédoniste2_postconf"]
                      ]
    
    key = value_selection[factor.index(key)]
    
    X = df[key].copy()
    X = add_constant(X)
    
    corr = df[key].corr()
    datatab = pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], index=X.columns)
    datatab = pd.DataFrame(datatab)
    datatab["index"] = datatab.index
    datacorr = pd.DataFrame(corr)
    datacorr["index"] = corr.index
    
    frame_filsx = Toplevel(root)
    ptres = Table(frame_filsx, dataframe=datatab )
    ptres.show()
    
    frame_fils = Toplevel(root)
    pt = Table(frame_fils, dataframe=datacorr)
    pt.show()
```

```python
import semopy as sem
import plotly.express as px
from skimage import io
from semopy import Model

pd.set_option('display.max_rows', 1000)  # or 1000
pd.options.display.float_format = '{:,.3f}'.format
# Confiance_marque =~ confiance_distrib_nAlim + confiance_distrib_alim + confiance_distrib + confiance_brico
# Image_marque =~ image_distrib_nAlim + image_distrib_alim + image_distrib + image_brico 
def model_print():
    desc= """
        # measurement model     
        cat_social =~ Diplome_FR + CSP + Cat_Revenue_FR
        Habitation =~ cat_Agglo + cat_Agglo_simple + cat_zone_habitation 
        
        Volonte_Ecologique =~ accord_achat_ecolo_postconf + intention_ecologie_postconf
        Volonte_Economie =~ accord_achat_econome_postconf + intention_economie_postconf + intention_economie2_postconf
        
        hedonisme =~ accord_consomation_hédoniste_postconf + accord_consomation_hédoniste2_postconf
        Legitimite =~ LEGIT_1 + LEGIT_2 + LEGIT_3 + LEGIT_4 + LEGIT_5 + LEGIT_6 + LEGIT_7 + LEGIT_9
        
        Trend_Local =~ TRENDS_LOC_1 + TRENDS_LOC_2 + TRENDS_LOC_3
        Trend_prix =~ TRENDS_PRIX_1 + TRENDS_PRIX_2 + TRENDS_PRIX_3
        Trend_secu =~ TRENDS_SECU_1 + TRENDS_SECU_2 + TRENDS_SECU_3 + TRENDS_SECU_4
        Trend_nouveaute =~ TRENDS_NEW_1 + TRENDS_NEW_2 + TRENDS_NEW_3
        
        confiance_comp =~ CONF_COMP_1 + CONF_COMP_2 + CONF_COMP_3
        confiance_benev =~ CONF_BENEV_1 + CONF_BENEV_2 + CONF_BENEV_3
        confiance_integ =~ CONF_INTEG_1 + CONF_INTEG_2 +  CONF_INTEG_3 + CONF_INTEG_4
        confiance =~ confiance_integ + confiance_benev + confiance_benev + confiance_comp
        
        image_val =~ IMG_VAL_1 + IMG_VAL_2 + IMG_VAL_3 + IMG_VAL_4 
        image_rse =~ IMG_RSE_1 + IMG_RSE_2 + IMG_RSE_3 + IMG_RSE_4 + IMG_RSE_5
        Image =~ image_rse + image_val
        
        Changement_Livraison =~ habitude_InternetTolivraison_postconf 
        Changement_Deplacement =~ habitude_InternetToPointvente_postconf + habitude_InternetToDrive_postconf + habitude_InternetTolocker_postconf
        Achat_Reconditionne =~ raison_achat_reconditionner_postconf + achat_magasin_reconditionne_postconf + achat_enligne_reconditionne_postconf
        
        # regressions
        Changement_Deplacement ~ Age + hedonisme + Achat_Reconditionne 
        Changement_Livraison ~ Age  + Image + hedonisme + Habitation
        Achat_Reconditionne ~ Volonte_Economie + hedonisme
        
        Image ~ confiance 
        confiance ~  Legitimite 
        Legitimite ~ Trend_secu + Trend_Local  + Trend_prix
        
        Trend_nouveaute ~ Image 
        
        Volonte_Economie ~ Volonte_Ecologique
        
        # residual correlations
        Trend_prix ~~ Trend_secu + Trend_nouveaute + Trend_Local
        Trend_nouveaute ~~ Trend_Local
        cat_social ~~ Habitation
        habitude_InternetTolivraison_postconf ~~ habitude_InternetToPointvente_postconf
        habitude_InternetTolivraison_postconf ~~ habitude_InternetToPointvente_postconf + habitude_InternetToDrive_postconf + habitude_InternetTolocker_postconf
        habitude_InternetTolivraison_postconf ~~ habitude_InternetToDrive_postconf
        habitude_InternetTolivraison_postconf ~~ habitude_InternetTolocker_postconf
    """

    model = Model(desc)
    res = model.fit(df)
    ins = model.inspect(mode='list', what="names", std_est=True)
    
    g = sem.semplot(model, "BatModel.jpg")
    
    frame_filsx = Toplevel(root)
    ptres = Table(frame_filsx, dataframe=sem.calc_stats(model))
    ptres.show()
    frame_fils = Toplevel(root)
    pt = Table(frame_fils, dataframe=ins[ins["op"] == "~"])
    pt.show()
    
    img = io.imread('BatModel.jpg')
    fig = plt.imshow(img)
    canvas = FigureCanvasTkAgg(fig,master = root)       
    canvas.draw() 
    canvas.get_tk_widget().pack() 
    toolbar = NavigationToolbar2Tk(canvas,root) 
    toolbar.update() 
    canvas.get_tk_widget().pack() 
```
