```python
import pandas as pd
import pyreadstat
import numpy as np
import json
pd.set_option('display.max_columns', None)

df_vague1 = pd.read_excel("/Users/fnacboulogne/Downloads/donnée_Quanti_avecMeta.xlsx", sheet_name="Vague 1")
df_vague2 = pd.read_excel("/Users/fnacboulogne/Downloads/donnée_Quanti_avecMeta.xlsx", sheet_name="Vague 2")
df_vague3 = pd.read_excel("/Users/fnacboulogne/Downloads/donnée_Quanti_avecMeta.xlsx", sheet_name="Vague 3")
df_col = pd.read_excel("/Users/fnacboulogne/Downloads/donnée_Quanti_avecMeta.xlsx", sheet_name="MetaColumn")
frame = [df_vague1,df_vague2, df_vague3]
df = pd.concat(frame)

#df = df.fillna(df.mean())
#df = df.fillna(0)
type(df)
```

```python
df.head()
```

```python
df_col = df_col.replace(' ', '_', regex=True)
```

```python
with open('/Users/fnacboulogne/Downloads/QuantiMetadata.json') as json_file:
    dict_metadata = json.load(json_file)
```

```python
df_meta = df.copy()
for columns in df:
    if columns in dict_metadata:
        df_meta[columns] = df_meta[columns].astype(str).map(dict_metadata[columns]).fillna(df[columns])
        #if pd.api.types.is_float_dtype(df_meta[columns]):
        #    df_meta[columns] = df_meta[columns].astype(str).map(dict_metadata[columns]).fillna(df[columns])
        #else :
        #    df_meta[columns] = df_meta[columns].astype(str).map(dict_metadata[columns]).fillna(df[columns])
```

```python
df_col.head()
```

```python
for columns in df:
    for head in df_col:
        if columns == df_col[head].iloc[0]:
            df.rename(columns={columns: df_col[head].iloc[1]},inplace=True)
            df_meta.rename(columns={columns: df_col[head].iloc[1]},inplace=True)
```
